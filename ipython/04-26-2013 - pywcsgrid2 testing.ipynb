```python
import pylab as plt

import matplotlib.path as mpath
import matplotlib.patches as mpatches

x = plt.arange(10)
y = plt.arange(10)

fig = plt.figure()
ax = fig.add_subplot(111)

path = [ (mpath.Path.MOVETO, [x[0],y[0]]) ]
for _x,_y in zip(x,y): 
    path.append( (mpath.Path.LINETO, [_x,_y]) )

codes, verts = zip(*path)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path)
ax.add_patch(patch)
```

```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches

verts = [
    (0., 0.), # left, bottom
    (0., 1.), # left, top
    (1., 1.), # right, top
    (1., 0.), # right, bottom
    (0., 0.), # ignored
    ]

codes = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY,
         ]

path = Path(verts, codes)

fig = plt.figure()
ax = fig.add_subplot(111)
patch = patches.PathPatch(path, facecolor='orange', lw=2)
ax.add_patch(patch)
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pyfits

import matplotlib.patheffects

fits_name = "/u/gl/lande/temp/lambda_mollweide_halpha_fwhm06_0512.fits"
f = pyfits.open(fits_name)
data = f[1].data
header = f[1].header

# used fixed header
del header["LONPOLE"]
del header["LATPOLE"]

fig = plt.figure(1, figsize=(8, 5))
fig.clf()

from pywcsgrid2.allsky_axes import make_allsky_axes_from_header
ax = make_allsky_axes_from_header(fig, rect=111, header=header,lon_center=0.)

data[:]=0
im = ax.imshow(data, origin="lower", cmap=plt.cm.gray_r)
im.set_clip_path(ax.patch)
im.set_clim(0, 30)

        
ax.grid()
plt.show()
```

```python
startx,starty=40,40
delt=10
x=[startx-delt, startx-delt, startx+delt, startx+delt]
y=[starty-delt, starty+delt, starty+delt, starty-delt]
path = [(mpath.Path.MOVETO, [x[0],y[0]])]
for _x,_y in zip(x,y): 
    path.append([mpath.Path.LINETO, [_x,_y]])
path.append([mpath.Path.CLOSEPOLY, [x[0], y[0]]]) 

codes, verts = zip(*path)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path, edgecolor='red', facecolor='orange')
ax['gal'].add_patch(patch)
fig
```
